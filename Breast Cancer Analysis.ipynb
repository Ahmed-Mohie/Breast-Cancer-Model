```python
# Loading helpful libraries
import numpy as np # Mathematical operations and linear algebra
import pandas as pd # Date exploration and processing
import matplotlib.pyplot as plt # Data visualization
import seaborn as sns # Data visualization

# Loading libraries for modeling
from sklearn.model_selection import train_test_split # Training data
from sklearn.preprocessing import StandardScaler # Standardization or normalization 
from sklearn.linear_model import LogisticRegression # classification model
from sklearn.metrics import accuracy_score # For accuracy testing
```

```python
# loading and readin data 
df = pd.read_csv('breast-cancer.csv')
df.head()
```

```python
# converting categorical values to numirecal values
df = df.copy()
df["diagnosis"]= df['diagnosis'].replace({'M' :1, 'B' : 0} )
```

```python
# Creatin standard scaler
scaler = StandardScaler()
```

```python
scaler
```

```python
# Define features and label
X = df.drop("diagnosis", axis = 1)
y = df["diagnosis"]
```

```python
# Fitting scaler to training data 
x_scaled = scaler.fit_transform(X)
```

```python
x_scaled
```

```python
y
```

```python
X
```

```python
# Splitting the data 
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.2, random_state = 0)
```

```python
x_train
```

```python
y_train
```

```python
# creating model
#log_reg = LogisticRegression( C=0.9,l1_ratio=1) (same accuracy)
# log_reg = LogisticRegression( intercept_scaling=2 (same accuracy)
log_reg = LogisticRegression()
```

```python
# Fitting the model
log_reg.fit(x_train, y_train)
```

```python
#calculating predictions
y_pred = log_reg.predict(x_test)
```

```python
# Chicking model accuracy
acc = accuracy_score(y_test, y_pred)
acc
```
